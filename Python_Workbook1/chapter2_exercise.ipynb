# EXERCISE **2**

Write a program to prompt the user for hours and rate per hour to compute gross pay.
```
Enter Hours: 35
Enter Rate: 2.75
Pay: 96.25
```

````python
#exercise 2
hrs = float(input("Enter Hours: "))
rate = float(input("Enter Rate: "))
pay = hrs * rate

print("Pay: ", pay)
````

# EXERCISE **4**

```
Enter a number: 3
```
