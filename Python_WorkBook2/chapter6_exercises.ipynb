Given the code below
```
text = 'Business Analytics'
```
What will the following expressions produce?
```
text[4]
text[2:4]
text[:4]
text[4:]
text[-2]
text[-4:-2]
text[-2:]
text[:-2]
```

````python
text = 'Business Analytics'
print(text)

letter = text[4]
print(letter)
letter1 = text[2:4]
print(letter1)
letter2 = text[:4]
print(letter2)
letter3 = text[4:]
print(letter3)
letter4 = text[-2]
print(letter4)
letter5 = text[-4: -2]
print(letter5)
letter6 = text[-2:]
print(letter6)
letter7 = text[:-2]
print(letter7)

````

# **Exercise 2**
Given the code below:
```
phrase = 'One Ring to rule them all'
```
What string slicing expression gives out the following words:
```
'One '
'Ring'
'all'
'them'
'One Ring to rule them all'
```

````python
phrase = 'One Ring to rule them all'

letter = phrase[:3]
print(letter)
letter1 = phrase[4:8]
print(letter1)
letter2 = phrase[-3:]
print(letter2)
letter3 = phrase[17:21]
print(letter3)

print(phrase)

````
