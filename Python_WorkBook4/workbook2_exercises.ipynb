```python
def sum(num1, num2):
  return num1 + num2

def diff(num1, num2):
  return num1 - num2

def multiply(num1, num2):
  return num1*num2

def divide(num1 , num2):
  if num2 == 0:
    print("error")
  return num1/num2

result_of_sum = sum(5, 3)
result_of_diff = diff(5, 3)
result_of_multiply = multiply(5, 3)
result_of_divide = divide(5, 3)

print("Sum:", result_of_sum)
print("Difference:", result_of_diff)
print("Multiply:", result_of_multiply)
print("Divide:", result_of_divide)
```
